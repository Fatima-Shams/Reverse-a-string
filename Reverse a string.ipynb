# Reverse a string

using for loop

```python
def reverse(s):
    result = ""
    
    
    for i in s:
        result = i + result
        """print(result)"""
    return result
    

s = "Shamshi Fatima"
print("original string:",s)
print("string after reversal:", reverse(s))
```

inbuilt reversed() method

```python
s = "Shamshi Fatima"
result = "".join(reversed(s))
print(result)
```

Extended slice indexing

```python
s = "Shamshi Fatima"
print(s[::-1])
```
